```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
RB = pd.read_csv('RBstats')
```

```python
def to_float(x):
    try:
        lst = x.split(',')
        final = ''
        for i in lst:
            final = final + i
        return float(final)
    except: 
        return 0
```

```python
RB['Yds'] = RB['Yds'].apply(to_float)
RB['Avg'] = RB['Avg'].apply(to_float) 
RB['1st%'] = RB['1st%'].apply(to_float)
RB['1st'] = RB['1st'].apply(to_float)
RB['20+'] = RB['20+'].apply(to_float)
RB['40+'] = RB['40+'].apply(to_float)
RB['FUM'] = RB['FUM'].apply(to_float)
```
